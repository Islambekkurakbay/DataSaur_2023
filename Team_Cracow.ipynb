```python
def preprocess_data(data, title='title', content='content'):
    '''
    the DataFrame data must have columns sys_lang, title, content
    '''
    data = data.dropna()
    data = data[data['sys_lang'].isin(['ru', 'kk', 'en'])]
    # script for detecting latin and cyrillic words
   
    kazakh_pattern = re.compile(r'[ӘәҒғҚқҢңӨөҰұҮүҺһіІ]')
    latin_pattern = re.compile(r'[a-zA-Z]')
    cyrillic_pattern = re.compile(r'[\u0400-\u04FF]')  # Covers the main block of Cyrillic characters

    def contains_kazakh(text):
        return bool(kazakh_pattern.search(text))    # Unique letters to Kazakh: Әә, Ғғ, Ққ, Ңң, Өө, Ұұ, Үү, Һһ
    def contains_latin(text):
        return bool(latin_pattern.search(text))
    def contains_cyrillic(text):
        return bool(cyrillic_pattern.search(text))
    
    series_contains_latin = data[title].apply(contains_latin)
    series_contains_cyrillic = data[title].apply(contains_cyrillic)
    
    # remove those rows that are for title column neither cyrillic, nor latin
    to_drop = data[~series_contains_latin & ~series_contains_cyrillic].index
    data = data.drop(to_drop)
    try: 
        series_contains_latin = data[content].apply(contains_latin)
        series_contains_cyrillic = data[content].apply(contains_cyrillic)
    except:
        pass
    
    # remove rows that are in content data neither cyrillic, nor latin (numbers etc.)
    to_drop = data[~series_contains_latin & ~series_contains_cyrillic].index
    data = data.drop(to_drop)
    
    # remove rows that are labeled 'en', but contain cyrillic content
    cyr = data[~series_contains_latin & series_contains_cyrillic]
    to_drop = cyr[cyr['sys_lang'] == 'en'].index
    data = data.drop(to_drop)
    
    # remove rows that are labeled 'ru or kk', but contain cyrillic content
    lat = data[series_contains_latin & ~series_contains_cyrillic]
    to_drop = lat[lat['sys_lang'].isin(['ru', 'kk'])].index
    data = data.drop(to_drop)

    # remove rows whose title contain kazakh language for non-kazakh sys_lang
    series_contains_kazakh = data[title].apply(contains_kazakh)
    kaz = data[series_contains_kazakh & series_contains_cyrillic]
    to_drop = kaz[~(kaz['sys_lang'] == 'kk')].index
    data = data.drop(to_drop)
    
    # not removed, because all russian letters also exist in kazakh alphabet
#     rus = articles[~series_contains_kazakh & series_contains_cyrillic & ~series_contains_latin]
#     rus[~(rus['sys_lang'] == 'ru')].head(3)

    return data
    
a = preprocess_data(pd.read_csv('articles.csv'))
```

```python
def kazakh_latin_to_cyrillic(text):
    # Define a dictionary for Kazakh Latin to Cyrillic conversion
    conversion = {
        'Zh': 'Ж', 'zh': 'ж',
        'Sh': 'Ш', 'sh': 'ш',
        'iy': 'йы', 'gi': 'ги',
        'ia': 'я', 'it': 'йт', 'qu': 'құ', 'Qu': 'Құ', 'ai': 'ай',
        'ii': 'йі', 'ie': 'ие', 'iy': 'иы', 'Zhu': 'Жұ', 'zhu': 'жұ',
        'A': 'А', 'a': 'а',
        'Á': 'Ә', 'á': 'ә',
        'B': 'Б', 'b': 'б',
        'D': 'Д', 'd': 'д',
        'E': 'Е', 'e': 'е',
        'F': 'Ф', 'f': 'ф',
        'G': 'Г', 'g': 'г',
        'Ǵ': 'Ғ', 'ǵ': 'ғ',
        'H': 'Х', 'h': 'х',
        'I': 'і', 'i': 'і',
        'U': 'Ы', 'u': 'ы',
        'J': 'Ж', 'j': 'ж',
        'K': 'К', 'k': 'к',
        'L': 'Л', 'l': 'л',
        'M': 'М', 'm': 'м',
        'N': 'Н', 'n': 'н',
        'Ń': 'Ң', 'ń': 'ң',
        'O': 'О', 'o': 'о',
        'Ó': 'Ө', 'ó': 'ө',
        'P': 'П', 'p': 'п',
        'Q': 'Қ', 'q': 'қ',
        'R': 'Р', 'r': 'р',
        'S': 'С', 's': 'с',
        'Ś': 'Ш', 'ś': 'ш',
        'T': 'Т', 't': 'т',
        'ý': 'у',
        'Ú': 'Ұ', 'ú': 'ұ',
        'Ū': 'Ү', 'ū': 'ү',
        'V': 'В', 'v': 'в',
        'Y': 'Ы', 'y': 'ы',
        'Z': 'З', 'z': 'з',
        'Ch': 'Ч', 'ch': 'ч',
        'Sh': 'Ш', 'sh': 'ш',
        'Ng': 'Ң', 'ng': 'ң',
        
    }

    # Replace each Latin letter with its corresponding Cyrillic letter
    for latin, cyrillic in conversion.items():
        text = text.replace(latin, cyrillic)
    return text
```

```python
import pandas as pd
import numpy as np
from gensim.models import FastText

# Create a list of sentences for FastText training
sentences = [title.split() for title in eng['title']]

# Train a FastText model on the sentences
model = FastText(sentences=sentences, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# Compute FastText embeddings for all titles
embeddings = np.array([model.wv.get_vector(title) for title in eng['title']])

def get_embedding(query):
    words = query.split()
    vectors = [model.wv[word] for word in words if word in model.wv]
    return np.mean(vectors, axis=0)

def find_most_similar_question(query, used_indices):
    query_vec = get_embedding(query[0])
    
    # Compute the cosine similarity
    dot_products = embeddings.dot(query_vec)
    norms_a = np.linalg.norm(embeddings, axis=1)
    norms_b = np.linalg.norm(query_vec)
    cosine_similarities = dot_products / (norms_a * norms_b)
    
    # Set similarity of already used titles to a very negative number
    cosine_similarities[used_indices] = -1e10
    
    # Get the index of the most similar question
    best_match_idx = cosine_similarities.argmax()
    
    return best_match_idx

# Test
used_indices = []
my_final = []

for i in test_latin['id']:
    result_idx = find_most_similar_question(test_latin[test_latin['id'] == i]['question'].to_list(), used_indices)
    used_indices.append(result_idx)
    my_final.append([i, eng.iloc[result_idx]['Unnamed: 0']])

my_final
```
